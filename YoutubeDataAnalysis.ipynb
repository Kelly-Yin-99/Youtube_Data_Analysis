```python
def convert_date(date_str):
    year = int(date_str[0:2]) + 2000  # Extracting and converting the year
    day = date_str[3:5]  # Extracting the day
    month = date_str[6:8]  # Extracting the month
    return pd.to_datetime(f"{year}-{month}-{day}")

# Apply the function to the 'trending_date' column
combined_df['trending_date'] = combined_df['trending_date'].apply(convert_date).dt.strftime('%Y-%m-%d')

combined_df['trending_date'].head()  # Displaying the first few converted dates for verification
```

### 2. Regression Analysis

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder



# Preparing the data with the filtered categories
X = combined_df[['likes','comment_count','category_id']]  # Using the specified variables
Y = combined_df['views']

# Reset the index of Y
Y = Y.reset_index(drop=True)

# Handling categorical data
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), ['likes','comment_count']),  # Scaling the 'likes' column
    ('cat', OneHotEncoder(), [ 'category_id'])  # OneHotEncoding the categorical columns
], remainder='passthrough')

# Transforming the data
X_transformed = preprocessor.fit_transform(X)

# Convert the transformed data to a dense array if it's sparse
if hasattr(X_transformed, "toarray"):
    X_transformed = X_transformed.toarray()

# Get feature names from OneHotEncoder and numeric features
onehot_features = list(preprocessor.named_transformers_['cat'].get_feature_names_out(['category_id']))
numeric_features = ['likes','comment_count' ]
feature_names = onehot_features + numeric_features

# Create DataFrame from transformed data
X_transformed_df = pd.DataFrame(X_transformed, columns=feature_names)

# Add constant for intercept
X_with_constant = sm.add_constant(X_transformed_df)

# Fit the model
model = sm.OLS(Y, X_with_constant).fit()

# Print the summary
model_summary = model.summary()
model_summary
```
